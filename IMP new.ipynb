```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
TVD = int(input("Enter Depth(ft)"))
MW = float(input("Enter mud weight(ppg)"))              
Kick = float(input("Enter  kick(ppg)"))                
ID = float(input("Enter casing ID(inch)"))          
OD = float(input("Enter drill string OD(inch)"))    
V = float(input("Enter kick volume(bbl)"))   
```

```python
df = pd.DataFrame({"Depth": range(TVD+1)})
df["HSP"] = 0.052*(MW)*(df["Depth"])              
df["FP"] = 0.052*(Kick)*(df["Depth"])              
df["SIDPP"] = df["FP"] - df["HSP"]                   
AC = ((ID**2)-(OD**2))/1029.4 
H = ((V*5.615)/(AC*5.615))
df["SICP"] = (df["FP"]-(0.052*H+0.052*12.5*(10000- H))) 
df.drop(0 ,axis=0)
```
